```python
# 定义需要编码的列
columns_to_check = [
    '当前终端品牌',
    '上次终端品牌',
    '上上次终端品牌',
    '家庭交往圈终端品牌偏好',
    '用户历史终端品牌偏好',
    '集团交往圈终端品牌偏好',
    '新5G终端品牌'
]

# 定义类别和对应整数的字典
category_mapping = {
    '华为': 1,
    '苹果': 2,
    'vivo': 3,
    'OPPO': 4,
    '荣耀': 5,
    '小米': 6,
    '三星': 7,
    '魅族': 8,
    '金立': 9,
    '中兴': 10,
    '一加': 11,
    '中国移动': 12,
        '其他': 13
}

# 循环处理每一列并进行编码
for column in columns_to_check:
    if column in df.columns:
        df[column] = df[column].replace(category_mapping)
df.head()
```

# 众数 填缺失值

```python
columns_with_missing_values = df.columns[df.isnull().any()].tolist()
columns_with_missing_values
```

```python
import pandas as pd

def fill_missing_with_mode(df):
    """
    遍历 DataFrame 的所有列，并使用众数填充缺失值
    Args:
        df: pandas DataFrame，需要处理的数据集

    Returns:
        filled_df: pandas DataFrame，填充缺失值后的数据集
    """
    filled_df = df.copy()

    # 遍历每列
    for col in filled_df.columns:
        # 获取当前列的众数
        mode_val = filled_df[col].mode()[0]

        # 使用众数填充缺失值
        filled_df[col].fillna(mode_val, inplace=True)

    return filled_df

# 调用函数填充缺失值
filled_df = fill_missing_with_mode(df)
```

```python
columns_with_missing_values = filled_df.columns[filled_df.isnull().any()].tolist()
columns_with_missing_values
```
